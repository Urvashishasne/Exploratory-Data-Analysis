```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['date']=data['Date_of_Journey'].str.split('/').str[0]
```

```python
data['month']=data['Date_of_Journey'].str.split('/').str[1]
```

```python
data['year']=data['Date_of_Journey'].str.split('/').str[2]
```

```python
data
```

```python
data['date']=data['date'].astype(int)
data['year']=data['year'].astype(int)
data['month']=data['month'].astype(int)
data.info()
```

```python
#data.drop('Date_of_Journey', axis=1, inplace=True)
data
```

```python
data['Dep_Time_hour']=data['Dep_Time'].str.split(':').str[0]
data['Dep_Time_min']=data['Dep_Time'].str.split(':').str[1]
```

```python
data.drop('Dep_Time', axis=1, inplace=True)
```

```python
data['Dep_Time_hour']=data['Dep_Time_hour'].astype(int)
data['Dep_Time_min']=data['Dep_Time_min'].astype(int)
data.info()
```

```python
data['Arrival_Time']=data['Arrival_Time'].apply(lambda x : x.split(' ')[0])
data.head()
data['Arrival_Time_hour']=data['Arrival_Time'].str.split(':').str[0]
data['Arrival_Time_min']=data['Arrival_Time'].str.split(':').str[1]
```

```python
data.info()
```

```python
data['Arrival_Time_hour']=data['Arrival_Time_hour'].astype(int)
data['Arrival_Time_min'] =data['Arrival_Time_min'].astype(int)
```
